```python
# All results with features from Aug 26:
dirname = 'opt-aug-26'

df = pd.DataFrame()
for filename in os.listdir(dirname):
    if filename.endswith('.csv'):
        path = os.path.join(dirname, filename)
        tmp_df = pd.read_csv(path)
        df = df.append(tmp_df)

len(df)
```
